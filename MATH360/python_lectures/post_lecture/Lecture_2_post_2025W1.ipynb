### Example - SIAM problem

Gas prices change on a nearly daily basis, and not every gas station offers the same price for a litre of gas. The gas station selling the cheapest gas may be across town from where you are driving. Is it worth the drive across town for less expensive gas? Create a mathematical model that can be used to help understand under what conditions it is worth the drive.

This problem is adapted from [link](https://m3challenge.siam.org/resources/archives/2016-year-at-a-glance/2016-problem-share-and-car-share-alike-modeling-new-approaches-to-mobility/) and [link](https://m3challenge.siam.org/what-is-math-modeling/modeling-handbooks/).

### Step 1: Identify the Problem

Problem: Which is the cheaper cost? Buying gas at the station on our route or going out of our way to a station that has a cheaper price.

### Step 2: Define Variables and Parameters

Variables:
- Let $P_A$ be the price of gas at Station A in dollars per litre.
- Let $P_B$ be the price of gas at Station B in dollars per litre.
- Let $k$ be the number of kilometres between the convenient gas station ("Station A") and the out-of-the-way station ("Station B").
- Let $S$ be the difference in cost between buying gas at Station B vs Station A in dollars.
    - here we are NOT including travel costs in $S.$
- Let $T$ be the cost to travel to and from Station B from Station A (i.e., a round trip from Station A to Station B).

Fixed Variables:
- Let $n$ be the number of litres of gas we buy.
- Let $f_e$ be the fuel economy of our vehicle in kilometers per litre.

Inputs: $k, P_A, P_B, n, f_e$

Outputs: $S, T$

### Step 3: Identify Assumptions and Constraints
Assumptions:
- There is a station out of our way that has cheaper gas
- We will buy a fixed amount of gas (is this realistic?)
- The car has a constant fuel economy

### Step 4: Build Solutions
Relationships:
We can write down two equations: one to describe the difference in gas station price and one to calculate the cost of going to Station B.

In going from Station A to Station B, note that we must travel $k$ km to Station B and then $k$ km back to Station A (i.e., where our route was). 

\begin{align*}
S & = \left(P_A - P_B\right)n \\
T & = \dfrac{2kP_B}{f_e}
\end{align*}

If $S>T$, then it is advantageous to go to Station B.  If $S\leq T$, then we should just go to Station A. 

We want to find the values for which $S=T$ or $\left(P_A - P_B\right)n=\dfrac{2kP_B}{f_e}.$  

In our model, we will have known values for $P_B$, $P_A$, $n$.  If we rearrange to find the distance $k$ that makes $S=T,$ we can then compare this with the actual distance between Station A and Station B to determine which scenario is optimal.  So we want to find $k = \dfrac{f_e\left(P_A - P_B\right)n}{2P_B}.$

Some values: 
- $P_B = \$ 1.92$ per litre
- $P_A = \$ 2.01$ per litre
- $n = 40$ litres
- $f_e = 100$ km per $5$ litres

```python
P_B = 1.92
P_A = 2.01
n = 40
f_e = 100/5

k = f_e * ((P_A-P_B)*n)/(2*P_B) 

print("Break even distance between Station A and B is", k, "km")
```

### Step 5: Analyze and Assess the Solution

Does this model do a good job?

Some diagnostic questions:
- Does the sign of the answer make sense? 
- Does the magnitude of the answer seem reasonable?
- Does the model change appropriately if we increase/decrease the input variables? 
- Are we correctly accounting for the distance between the Stations?

Other questions to consider:
- Are the assumptions reasonable and defensible?
- Are the assumptions relevant?
- Is the model explained by our assumptions or do we need to enforce more assumptions?
- Are our assumptions correct? 
- Are our results scenario dependent? (i.e., do our results change a lot or a little when we make a small change in the scenario?)

#### Iterate
This is often tied into Step 5, where we can make improvements to the model based on our critical analysis.

Perhaps we want to make some improvements to the model:
- We could track both cost and distance.
- We might want to consider making the number of litres of gas to buy $n$ a function of distance.
- We may want to include a non-constant fuel economy.
- We may want to include a factor to account for how willing someone would be to go off their planned route. E.g., some drivers may not want to drive significantly out of their way to save a "minimal" amount of money.  Some will be okay with paying more for the convenience factor.  A more complex model may want to look at how this aspect influence the model.
- We may want to consider the time that it takes to go off the main route in the model as a "cost" in addition to the monetary cost of the gas.
- Others?

These are all improvements that we can adjust in the model to refine our solution. We may find that to make improvements, we will need to adjust our variables and parameters, our assumptions and constraints, and how the solution is built. Once we are satisfied, then we move to step 6.

### Example -- Ski Trip
You and your friends are on a ski trip together in the mountains.  Just after arriving to your log cabin you get stuck in a snow storm that lasts a few days.

When the storm is over, you open the cabin door and realize that there is 2 metres of snow outside.  Unfortunately, you can only leave the cabin when the snow depth is 0.5 metres or less.  How long will you and your group be stuck in the cabin.

You and your group know the following information:
- Snow is constantly falling with a rate of $s$ metres per day
- Snow has just started to melt and the melting rate is linearly increasing by $m$ metres per day per day (i.e., units of m/(day)$^2$)
- Meltwater runs down the mountain (lowering the snow depth in front of your cabin), however 50% of the melted snow per day refreezes at the site. 

Other notes: 
- Ignore the compaction of the snow and other processes
- All units are in meters of water equivalent
- Assume that refreezing is happening at the same time as melting
- You can assume that snow is falling at a rate of 10 cm/day and that the melting rate is 2 cm/(day$^2$).

### Step 1: Identify the Problem

Problem: We want to model the snow depth outside to determine how long we will be stuck in the cabin.

### Step 2: Define Variables and Parameters
Let: 
- $s$ be the rate of fallin snow (m/day)
- $m$ be the snow melting rate (m/(day)$^2$)
- $t$ be time in days
- $D$ be the depth of the snow in m
- $R$ be the amount of refreezing water in m
- $S$ be the total snowfall rate in m/day
- $M$ be the total melting rate in m/day

### Step 3: Identify Assumptions and Constraints
- No snow compaction or other environmental processes.
- Refreezing occurs at the same time as melting.
- Snow depth is homogeneous (i.e., we can consider the snow behaves identically for every vertical slice).

### Step 4: Build Solutions

We want to create a model that includes the falling snow, the melting of snow, and the refreezing of water into the calculation for the depth of the snow. 

#### Snowfall
We are told there is a constant snowfall rate, so $S = s.$ 

#### Melting 
The rate of melting is linearly increased by $m$, so the melt rate is $M = mt.$

#### Refreezing
We know $50\%$ of the melted snow refreezes at the site.  The melt rate is $M=mt$, so the refreezing rate is $R = 0.5M \implies R = 0.5mt.$ 

#### Depth

The rate of change of the depth is the difference between "added" depth and "subtracted" depth. We can think about this problem in a similar way to a tank problem in differential equations: the rate of change is the in flow minus the out flow.

In the case of this problem, we have Snowfall and Refreezing "adding" to the depth and Melting "subtracting" from the depth.

\begin{align*} 
\frac{d D}{d t} &= ( S + R) - M \\
& = (s + 0.5mt) - mt\\
& = s - 0.5mt
\end{align*}

##### Additional information
We know $s =0.1$ and $m=0.02$ from the initial information, as well as that $D(0) = 2$.

This equation is a first order differential equation, which can be solved by separation of variables!

\begin{align*} 
\frac{d D}{d t} & = s - \frac{1}{2}mt \\
\implies D(t) & = st- \frac{1}{4}mt^2 + C \bigg|_0^{t=t^*}
\end{align*}

Inputing $D(0) = 2$ gives and relabelling $t^*$ with $t$:
$$D(t) = st - \frac{1}{4}mt^2 + 2.$$

```python
t = np.linspace(0,35,1000)

s = 0.1
m = 0.02
D = s*t-1/4*m*t**2+2

plt.plot(t,D)
plt.axhline(y=0.5, color='r', linestyle='--') #horizontal line for depth of 0.5m
plt.xlabel('Days')
plt.ylabel('Snow Depth')
plt.grid()
plt.show()
```

We want the intersection of $D = st-\frac{1}{4}mt^2+2$ and $d = 0.5$ to determine the number of days we will be in the cabin.  From the plot, it looks as thought $t \approx 30$, but let's find the exact value.

\begin{align*}
0.5 & = 0.1t-\frac{1}{4}(0.02)t^2+2\\
0 & = -\frac{1}{200}t^2 + 0.1t +1.5\\
\implies t & = \frac{-0.1 \pm \sqrt{(0.1)^2 - 4 \left( -\frac{1}{200} \right)(1.5)}}{2\left(-\frac{1}{200} \right)}\\
& = \frac{-0.1 \pm 0.2}{-\frac{1}{100}}\\
\end{align*}

So $t = -10$ days or $t = 30$ days.  

Therefore, we will need to wait $30$ days in the cabin before we are able to leave. 

### Step 5: Analyze and Assess the Solution

The solution seems to make sense: the rate of snowfall and refreezing (the inflow) is much larger than the rate of melting, so it will take a while for the snow to reach a depth of $0.5$ m so that we are able to leave the cabin. 

Ploting a range of $s$ and $m$ values below shows that behaviour seems to be correct.  As the rate of snowfall increases to $10$ times the given value, we see that it will take much longer for the depth to reach $0.5$ m, and as the rate is decreased by a factor of $10$, we see that it takes less time for the depth to reach $0.5$m.  We see a similar result for $m$: when the melting rate is increased by a factor of $10$, we reach a depth of $0.5$ m quicker, and when the melt rate is decreased by a factor of $10$, it takes much longer to reach a depth of $0.5$ m.  When the refreezing amount is increased by a facotr of $2$, we see that the solution curve is increasing and there is no avenue for a depth of $0.5$ m to be reached without intervention.  When the refreezing amount is decreased by a factor of $10$, we see that attaining a depth of $0.5$ m occurs quicker than the original problem statement.  These analyses seem to be physically relevant and correct. 

Here, the result is dependent on the rate of melting, snowfall, and refreezing, so alterations to $s, m$, and $r$ will affect the output of the model. 

Other questions to consider:
- How does the refreezing rate change the output of the model?
- Is it reasonable to assume that there is no snow compaction and that refreezing occurs at the same time as melting?
- We may want to include efforts by the people in the cabin to lower the snow depth near the door.
- Is the assumption that the snow behaves homogeneously an okay assumption? E.g., Areas that are shaded will likely not melt/accumulate snow at the same rate as those in the open.

Note: in the plot below, when we are analyzing the solution by varying some of the inputs, the depth becomes negative, which is not physical.  In this step, we are looking at the general functional behaviour to assess if our solution makes sense.  The negative depth can be an indication that our solution is incorrect, so we do want to note this.  Here, the negative depth is related to the functional form of the solution (i.e., a quadratic function), so we are okay.

```python
t = np.linspace(0,35,1000)

s = 0.1
m = 0.02
D = s*t-1/4*m*t**2+2

plt.plot(t,s*t-1/4*m*t**2+2,label='Problem statement')
plt.plot(t,10*s*t-1/4*m*t**2+2,label='10s',linestyle='dotted')
plt.plot(t,0.1*s*t-1/4*m*t**2+2,label='0.1s',linestyle='dotted')
plt.plot(t,s*t-1/4*10*m*t**2+2,label='10m',linestyle='dashed')
plt.plot(t,s*t-1/4*0.1*m*t**2+2,label='0.1m',linestyle='dashed')
plt.xlabel('Days')
plt.ylabel('Snow Depth')
plt.legend()
plt.grid()
plt.show()
```

Varying the refreezing rate is a bit tickier, as it was collapsed into the $0.5mt$ term.  Let's go back to and introduce a parameter $r$ as the refreezing rate:
\begin{align*} 
\frac{d D}{d t} &= ( S + R) - M \\
& = (s + rmt) - mt\\
& = s + (r-1)mt
\end{align*}

Then the solution is:

\begin{align*} 
\frac{d D}{d t} & = s + (r-1)mt \\
\implies D(t) & = st + \frac{r-1}{2}mt^2 + C \bigg|_0^{t=t^*}
\end{align*}

Inputing $D(0) = 2$ gives and relabelling $t^*$ with $t$:
$$D(t) = st + \frac{r-1}{2}mt^2 + 2.$$

```python
s = 0.1
m = 0.02
r = 0.5

plt.plot(t,s*t+(r-1)/2*m*t**2+2,label='Problem statement')
plt.plot(t,s*t+(2*r-1)/2*m*t**2+2,label='2r',linestyle='dashed')
plt.plot(t,s*t+(0.1*r-1)/2*m*t**2+2,label='0.1r',linestyle='dotted')
plt.xlabel('Days')
plt.ylabel('Snow Depth')
plt.legend()
plt.grid()
plt.show()
```

By varying the refreezing rate, we seem to have physical results.  If the refreezing rate is too high, then we will not be able to leave the cabin, and if the refreezing rate is low, then we will be able to leave earlier. 

Other questions to consider:
- Does the "shape" of the solution make sense? I.e., is it reasonable for the $2r$ line to look more linear, while the $0.1r$ line looks more parabolic?
- Should the graphs for the problem statement and $0.1r$ be concave down?
